```python
import pandas as pd
import glob
```

```python
# Define the path to your files
path = r'../data\NY'

# Use glob to get all the CSV files in the folder
all_files = glob.glob(path + "/*.csv")

# Load all the files into a single DataFrame
df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)

# Verify the first records of the DataFrame
df.head()
```

```python
# Identify columns with missing values
missing_values = df.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

```python
# Remove rows with null values in 'end_station_id'
df = df.dropna(subset=['end_station_id'])

# Verify the rows have been removed
missing_values_after_removal = df.isnull().sum()

# Display columns with remaining missing values
missing_values_after_removal[missing_values_after_removal > 0]
```

```python
# Remove rows with null values in 'start_station_id'
df = df.dropna(subset=['start_station_id'])

# Verify the rows have been removed
missing_values_final = df.isnull().sum()

# Display columns with remaining missing values
missing_values_final[missing_values_final > 0]
```

```python
# Convert the 'started_at' and 'ended_at' columns to datetime objects
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# Display information about the DataFrame
df.info()
```

```python
# Identify rows where the trip duration is negative
negative_duration = df[df['ended_at'] <= df['started_at']]

# Display the rows with negative trip duration
negative_duration
```

```python
# Adjust the 'ended_at' timestamps to correct negative durations caused by daylight saving time changes
df.loc[df['ended_at'] <= df['started_at'], 'ended_at'] += pd.Timedelta(hours=1)

# Verify the adjustment
negative_duration_corrected = df[df['ended_at'] <= df['started_at']]
negative_duration_corrected
```

```python
# Calculate trip duration in days
df['trip_duration_days'] = (df['ended_at'] - df['started_at']).dt.days

# Calculate trip duration in hours, excluding days
df['trip_duration_hours'] = ((df['ended_at'] - df['started_at']).dt.total_seconds() // 3600) % 24

# Calculate trip duration in minutes, excluding hours
df['trip_duration_minutes'] = ((df['ended_at'] - df['started_at']).dt.total_seconds() // 60) % 60

# Display the first few rows to inspect the new columns
df.head()
```

```python
# Convert 'started_at' and 'ended_at' columns to datetime objects
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# Display DataFrame information to verify conversion
df.info()
```

```python
# Add columns for day, month, and year
df['day'] = df['started_at'].dt.weekday
df['month'] = df['started_at'].dt.month
df['year'] = df['started_at'].dt.year

# Check the new columns
df[['day', 'month', 'year']].head()
```

```python
# Check that the station coordinates are within a valid range
invalid_start_coords = df[(df['start_lat'] < -90) | (df['start_lat'] > 90) | (df['start_lng'] < -180) | (df['start_lng'] > 180)]
invalid_end_coords = df[(df['end_lat'] < -90) | (df['end_lat'] > 90) | (df['end_lng'] < -180) | (df['end_lng'] > 180)]

# Display invalid coordinates
invalid_start_coords, invalid_end_coords
```

```python
# Export the cleaned data to a CSV file in the same path
df.to_csv('../data/cleaned_citibike_data_ny.csv', index=False)
```
